## **3.** Basic dataframe operations and visualization for the `Weather Data`

**3.1** Import the excel file `Excel_sample.xlsx` as a dataframe with name `df3` using pandas package

Sheet name is `Weather`

```python
#Data Visualization of weather data
file_name= "/Users/yachu/Downloads/e.xlsx"
df3=pd.read_excel(file_name,sheet_name='ewewe')
```

**3.3** Plot a histogram of the mean tempature (using the column `Mean Temp (°C)`). Set the `bins = 20`

```python
#Histogram of Min and Max
df3['Mean Temp (°C)'].plot.hist(bins=20, edgecolor='red', color='blue')
plt.title('Mean temperature')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for min,max and mean temperature
df3[['Min Temp (°C)','Max Temp (°C)','Mean Temp (°C)']].boxplot()
plt.title('Box Plot')
plt.ylabel('Temperature (°C)')
plt.xlabel('Categories')
plt.show()
```

```python
#Density plot
df3[['Min Temp (°C)','Max Temp (°C)','Mean Temp (°C)']].plot.density()
plt.title('Plot Density')
plt.xlabel('Temperature (°C)')
plt.show()
```
